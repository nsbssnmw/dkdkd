```python
# 1. 판다스 라이브러리 불러오기
import pandas as pd
```

```python
# 2. 데이터 프레임 만들기
df = pd.DataFrame(              # df라는 변수에 데이터 프레임을 담아 줍니다.
        {"a" : [24 ,25, 26, 27, 28],    # 열 이름을 지정해 주고 시리즈 형태로 데이터를 저장합니다.
        "b" : [29, 30, 31, 32, 33],
        "c" : [34, 35, 36, 37, 38],
        "d" : [39, 40, 41, 42, 43]},
        index = [21,22, 23, 24, 25])   # 인덱스는 1,2,3으로 정해 줍니다.
```

```python
# 3. 데이터 프레임 출력
df
```

```python
# 4. 데이터의 열 이름을 따로 지정해서 만들기
df = pd.DataFrame(
        [[24, 29, 34, 39],
        [25, 30, 35, 40],
        [26, 31, 36, 41],
        [27, 32, 37, 42],
        [28, 33, 38, 43]],
        index=[21, 22, 23, 24, 25],
        columns=['a', 'b', 'c', 'd'])  # 열 이름을 따로 정해줄 수 있습니다.
df
```

```python
# 5. 인덱스가 두 개인 데이터 프레임 만들기
df = pd.DataFrame(
    {"a" : [24 ,25, 26, 27, 28],
     "b" : [29, 30, 31, 32, 33],
     "c" : [34, 35, 36, 37, 38],
     "d" : [39, 40, 41, 42, 43],},
    index = pd.MultiIndex.from_tuples(      # 인덱스를 튜플로 지정합니다.
        [('g', 1), ('g', 2), ('h', 1), ('h', 2),('h', 3)],
        names=['q', 'w']))              # 인덱스 이름을 지정합니다.
df
```

```python
# 6. 특정 열의 값을 기준으로 정렬하기
df.sort_values('d', ascending=False)  # ascending=False를 적어주면 역순으로 정렬합니다.
```

```python
# 7. 열 이름 변경하기
df.rename(columns = {'d':'f'})  #c 열의 이름을 d로 변경합니다.
```

```python
# 8. 인덱스 값 초기화하기
df.reset_index()
```

```python
# 9. 인덱스 순서대로 정렬하기
df.sort_index()
```

```python
# 10. 특정 열 제거하기
df.drop(columns=['a', 'c'])
```

```python
# 11. 맨 위의 행 출력하기
df.head(3)  # 2행을 출력합니다.
```

```python
# 12. 맨 아래 행 출력하기
df.tail(3)  # 2행을 출력합니다.
```

```python
# 13. 특정 열의 값을 통해 추출하기
df[df["d"] > 25]          # a열 중 4보다 큰 값이 있을 경우 해당 행을 추출합니다.
```

```python
# 14. 특정 열에 특정 값이 있을 경우 추출하기
df[df["d"] == 40]       # a열 중 6이 있을 경우 해당 행을 추출합니다.
```

```python
# 15. 특정 열에 특정 값이 없을 경우 추출하기
df[df["d"] != 5]        # a열 중 5가 없을 경우 해당 행을 추출합니다.
```

```python
# 16. 특정 열에 특정 숫자가 있는지 확인하기
df[df['d'].isin([39])]  # 원하는 숫자를 리스트([int]) 형식으로 써줍니다.
```

```python
# 17. 특정 비율로 데이터 샘플링하기
df.sample(frac=0.7)  # 실행할 때마다 정해진 비율 만큼 랜덤하게 추출합니다.
```

```python
# 18. 특정 개수 만큼 데이터 샘플링하기
df.sample(n=4)  # 실행할 때마다 n에서 정한 만큼 랜덤하게 추출합니다.
```

```python
# 19. 특정 열에서 큰 순서대로 불러오기
df.nlargest(2,'d')   # a열에서 큰 순서대로 3개를 불러와 보겠습니다.
```

```python
# 20. 특정 열에서 작은 순서대로 불러오기
df.nsmallest(1,'c')   # a열에서 작은 순서대로 3개를 불러와 보겠습니다.
```

```python
# 21.인덱스의 범위로 불러오기

# 0부터 세므로 첫 번째 줄은 인덱스 0, 4째 줄은 인덱스 3이 됩니다.
df.iloc[0:3]    # [a:b]의 경우 a 인덱스부터 b-1인덱스까지 불러오라는 의미입니다.# a열에서 큰 순서대로 3개를 불러와 보겠습니다.
```

```python
# 22.첫 인덱스를 지정해 불러오기
df.iloc[1:]    # [a:]는 a인덱스부터 마지막 인덱스까지 불러오라는 의미입니다.
```

```python
# 23.마지막 인덱스를 지정해 불러오기
df.iloc[:4]    # [:b]는 처음부터 b-1 인덱스까지 불러오라는 의미입니다.
```

```python
# 24.모든 인덱스를 불러오기
df.iloc[:]    # [:]는 모든 인덱스를 불러오라는 의미입니다.
```

```python
# 25.특정 열을 지정하여 가져오기
df[['c', 'd']]   # a열과 b열을 가져오라는 의미입니다.
```

```python
# 26.조건을 만족하는 열 가져오기
df.filter(regex='d')    # 열 이름에 c라는 문자가 포함되어 있으면 출력하라는 의미입니다.
```

```python
# 27.특정 문자가 포함되지 않는 열 가져오기
df.filter(regex='^(?!d$).*')    # 열 이름에 c라는 문자가 포함되어 있지 않으면 출력하라는 의미입니다.
```

```python
# 28.특정한 행과 열을 지정해 가져오기
#df.loc[가져올 행,가져올 열]의 형태로 불러옵니다.
df.loc[:, 'c':'d']     # 모든 인덱스에서, a열부터 c열까지를 가져오라는 의미입니다.
```

```python
# 29. 인덱스로 특정 행과 열 가져오기
df.iloc[1:2, [1, 2]]   # 0 인덱스부터 2인덱스까지, 0번째 열과 2번째 열을 가져오라는 의미입니다. (첫 열이 0번째입니다.)
```

```python
# 30. 특정 열에서 조건을 만족하는 행과 열 가져오기
df.loc[df['b'] > 25, ['c', 'd']]  # a열의 값이 5보다 큰 경우의 a열과 c열을 출력하라는 의미입니다.
```

```python
# 31. 인덱스를 이용해 특정 조건을 만족하는 값 불러오기
df.iat[2, 2]    # 1번째 인덱스에서 2번째 열 값을 가져옵니다.
```
