```python
# 58. 모든 열을 행으로 변환하기
pd.melt(df)
```

```python
# 60. 여러 개의 열을 행으로 이동시키기
df_melt = pd.melt(df, id_vars=['B'], value_vars=['C','D'])    # A열만 그대로, B열과 C열은 행으로 이동시키기
df_melt
```

```python
# 실습을 위한 데이터 프레임 만들기 1
adf = pd.DataFrame({"x2" : ["A","B","C"],
                    "x1": [1,2,3],
                    "x3": [4,5,6]})
adf
```

```python
# 데이터 프레임 만들기 2
bdf = pd.DataFrame({"x2" : ["A","B","D"],
                    "x1": ["T","F","T"],
                    "x3": ["G","H","O"]})
bdf
```

```python
# 데이터 프레임 만들기 3
cdf = pd.DataFrame({"x2" : ["B","C","D"],
                    "x1": [4,5,6],
                    "x3": [6,7,8]})
cdf
```

```python
# 70. 왼쪽 열을 축으로 병합하기
pd.merge(adf, bdf, how='left', on='x2')
# x1을 키로 해서 병합, 왼쪽(adf)를 기준으로
# 왼쪽의 adf에는 D가 없으므로 해당 값은 NaN으로 변환
```

```python
# 71. 오른쪽 열을 축으로 병합하기
pd.merge(adf, bdf, how='right', on='x2')
# x1을 키로 해서 병합, 오른쪽(bdf)을 기준으로
# 오른쪽의 bdf에는 C가 없으므로 해당 값은 NaN으로 변환
```

```python
# 72. 공통의 값만 병합
pd.merge(adf, bdf, how='inner', on='x2')
```

```python
# 73. 모든 값을 병합
pd.merge(adf, bdf, how='outer', on='x2')
```

```python
# 74. 특정한 열을 비교하여 공통 값이 존재하는 경우만 가져오기
adf[adf.x1.isin(bdf.x2)]
# adf와 bdf의 특정한 열을 비교하여 공통 값이 존재하는 경우만 가져오기
# adf.x1열과 bdf.x1열은 A,B가 같다. 따라서 adf의 해당 값만 출력
```

```python
# 75. 공통 값이 존재하는 경우 해당 값을 제외하고 병합하기
adf[~adf.x1.isin(bdf.x2)]  # 해당 값만 빼고 출력
```

```python
# 76. 공통의 값이 있는 것만 병합
pd.merge(adf, cdf)
```

```python
# 77. 모두 병합
pd.merge(adf, cdf, how='outer')
```

```python
# 78. 어디서 병합되었는지 표시하기
pd.merge(adf, cdf, how='outer', indicator=True)
```

```python
# 79. 원하는 병합 만 남기기
pd.merge(adf, cdf, how='outer', indicator=True).query('_merge == "left_only"')
```

```python
# 80. merge 컬럼 없애기
pd.merge(adf, cdf, how='outer', indicator=True).query('_merge == "left_only"').drop(columns=['_merge'])
```
